#### **1.2. Convertidor de números romanos**

Los números romanos expresan números con 7 símbolos diferentes.

La tabla muestra el valor que representa cada símbolo.

Los números romanos se calculan fácilmente si agrega los símbolos en orden.
Por ejemplo, III es la suma de 1 por tres veces y XVI es la suma de 10, 5 y 1.
Sin embargo, si un símbolo que va detrás es un número mayor, resta el valor del símbolo que va delante.
Por ejemplo, IX es suma de 10 y resta 1, mientras que IV es suma de 5 y resta 1.

num rom1.PNG

¿Cómo harías para convertir los números romanos MDCLXVI?

Como se muestra a continuación, convierta cada símbolo en números arábigos y luego súmelos.

num rom2.PNG

```python
table={'I':1,'V':5,'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
def roman_to_int(str):
  result=0
  for i in range(len(str)-1):
    if table[str[i]]< table[str[i+1]]:
      result-=table[str[i]]
    else:
      result+=table[str[i]]
  return result+table[str[-1]]
```

```python
roman=input('Ingrese un número romano: ')
number=roman_to_int(roman)
print(number)
```

#### **2.1. Definición de tabla hash**

La tabla hash o el mapa hash se refiere a un tipo de datos abstractos que almacena y extrae datos asignando claves a valores a través de hashing.

Hashing se refiere a la designación de ubicaciones para guardar datos o encontrar datos guardados por valores de datos. La función hash determina la clave utilizando el valor de los datos.

La tabla hash debe proporcionar tres interfaces significativas de la siguiente manera.

- put (Key, value): Almacena el valor en el depósito o contenedor (bucket) de objetos designado por el valor hash clave.
- get (key): Extrae el valor almacenado en el depósito designado por el valor hash clave.
- hash (key): Convierte el valor de la clave de datos en un valor hash en un rango específico.

#### **2.3. Hashing y funciones hash**

Hashing se refiere a un proceso de convertir una clave dada en otra clave, y la función hash se refiere a una función que convierte una clave de longitud aleatoria en otra clave que tiene una longitud fija. La clave convertida por la función hash se llama clave hash.

Hay 8 ranuras o divisiones en la estantería, así que defina la función hash como $f(x) = x / 8$.

hash4.PNG

#### **2.4. Definir una tabla hash como un tipo de clase**

La definición de la tabla hash como una clase es la siguiente.

```python
class HashTable:

  def __init__(self,size):
    self.size=size
    self.table={}
    for i in range(size):
      self.table[i]=[]
    
  def hash(self,key):
    return key%self.size
  
  def get(self,key):
    return self.table[self.hash(key)]

  def put(self,key,value):
    bucket=self.table[self.hash(key)]
    if value not in bucket:
      bucket.append(value)
```

**Líneas 3-18:**

El constructor de la tabla hash inicializa la tabla hash en el tamaño de la clave dada.

- 3-7: El 'size' como parámetro de entrada es el tamaño de la tabla hash, y la tabla hash está definida por el diccionario de Python.
En el diccionario, inicialice el deposito (bucket) en número de tamaño. Al principio, el bucket se convierte en una lista vacía.

En la tabla hash, defina la función hash como una función que realiza aritmética modular con el tamaño de la clave dada.

El método get() recibe la clave como entrada y devuelve el bucket de clave hash.


El método put() recibe la clave y el valor como entrada y almacena el valor en el bucket de clave hash.

- 12-13: Devuelve el bucket donde el valor clave es self.hash(key) en el diccionario self.table.

- 15-18: Encuentre el bucket donde el valor clave es self.hash(key) en el diccionario self.table.
Cuando no se permite la duplicidad, agregue el valor en el bucket si el depósito no tiene el valor.

#### **2.5. Ejemplo de uso de la tabla hash**

Inicialice la tabla hash que se definió previamente en una tabla con 8 buckets.

```python
table=HashTable(8)
for key in table.table.keys():
  print(key,table.table[key])
```

#### **3.2. Implementación y codificación**

La conversión de números romanos se puede implementar de la siguiente manera usando la tabla hash:

```python
def roman_to_int(str):
  result=0
  for i in range(len(str)-1):
    if table[str[i]]< table[str[i+1]]:
      result-=table[str[i]]
    else:
      result+=table[str[i]]
  return result+table[str[-1]]
```

Líneas 1-3:

- 1-3: La función roman_to_int() convierte el str, que es un parámetro de entrada, en números arábigos.
El valor del resultado se inicializa a 0 al principio. Realiza sumas o restas a todas las letras.
- 4-7: Si el valor de la letra que está detrás es mayor que el valor de la letra actual, réstelo del valor actual.
Si no, agregue el valor de la letra al valor actual.
- 8: Al romper el bucle for, todas las letras, excepto la última, se suman o se restan.
Entonces, devuelve la suma de los valores de str[-1] (la última letra) y el resultado.

**Directriz, mecanismos y plan de contingencia**

 Preparar la programación por parejas implica establecer pautas y mecanismos para ayudar a los alumnos a emparejarse correctamente y mantenerlos emparejados.  Por ejemplo, los estudiantes deben turnarse para "conducir el mouse".  La preparación efectiva requiere planes de contingencia en caso de que uno de los socios se ausente o decida no participar por una u otra razón.  En estos casos, es importante dejar claro que el alumno activo no será sancionado porque el emparejamiento no funcionó bien.
 
**Emparejar habilidades similares, no necesariamente iguales, como socios**

 La programación en parejas puede ser efectiva cuando los estudiantes con habilidades similares, aunque no necesariamente iguales, se emparejan como compañeros.  El emparejamiento de estudiantes que no coinciden a menudo puede conducir a una participación desequilibrada.  Los maestros deben enfatizar que la programación en parejas no es una estrategia de "divide y vencerás", sino más bien un verdadero esfuerzo de colaboración en cada esfuerzo por todo el proyecto.  Los maestros deben evitar emparejar a estudiantes muy débiles con estudiantes muy fuertes.

**Motivar a los estudiantes ofreciendo incentivos adicionales**

 Ofrecer incentivos adicionales puede ayudar a motivar a los estudiantes a emparejarse, especialmente con estudiantes avanzados.  A algunos maestros les ha resultado útil pedirles a los estudiantes que se emparejen solo para una o dos tareas.

**Evitar trampas en la colaboración**

 El desafío para el maestro es encontrar formas de evaluar los resultados individuales, mientras aprovecha los beneficios de la colaboración.  ¿Cómo saber si un estudiante aprendió o hizo trampa?  Los expertos recomiendan revisar el diseño y la evaluación del curso, así como discutir explícita y concretamente con los estudiantes los comportamientos que se interpretarán como trampa.  Los expertos alientan a los maestros a hacer que las tareas sean significativas para los estudiantes y a explicar el valor de lo que los estudiantes aprenderán al completarlas.

**Entorno de aprendizaje colaborativo**

 Un entorno de aprendizaje colaborativo ocurre cada vez que un instructor requiere que los estudiantes trabajen juntos en actividades de aprendizaje.  Los entornos de aprendizaje colaborativo pueden implicar actividades tanto formales como informales y pueden incluir o no una evaluación directa.  Por ejemplo, parejas de estudiantes trabajan en asignaciones de programación; pequeños grupos de estudiantes discuten posibles respuestas a la pregunta de un profesor durante la lección; y los estudiantes trabajan juntos fuera de clase para aprender nuevos conceptos.  El aprendizaje colaborativo es distinto de los proyectos en los que los estudiantes "dividen y vencerán".  Cuando los estudiantes dividen el trabajo, cada uno es responsable de solo una parte de la resolución del problema y hay oportunidades muy limitadas para resolver los problemas con los demás.  En entornos colaborativos, los estudiantes participan en conversaciones intelectuales entre ellos.
