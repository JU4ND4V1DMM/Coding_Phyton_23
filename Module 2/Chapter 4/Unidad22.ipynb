#### **3.1. ¿Cómo funciona el corrector de paréntesis equilibrado?**

Las entradas de paréntesis equilibrados se clasifican en válidas o no válidas.

```python
def check (expr):
  opening= '[{('
  closing=']})'
  stack=[]
  out_result = True
  for char in expr:
    if char in opening:
      stack.append(char)
    elif char in closing:
      if opening.index(stack.pop()) != closing.index(char):
        out_result = False
  if len(stack) != 0:
    out_result = False
  return out_result
```

#### **5.1. Definición de una pila**

Stack es un tipo de datos abstracto en el que un elemento se puede agregar o eliminar solo en un lado de la parte superior. Este tipo de datos funciona con el método LIFO (Last-In-First-Out) en el que puede eliminar el elemento que se ingresó en el último lugar.

**Stack debe proporcionar las siguientes dos interfaces.**

- Push (apilar): Agrega un nuevo elemento en la parte superior.
- Pop (desapilar): Elimina un elemento de la parte superior.

#### **6.2. Definir una pila como un tipo clase**

En Python, el tipo de datos abstracto se puede definir como una clase.
El tipo de datos de pila abstracta que se implementó como una lista se puede definir como una clase, como se muestra a continuación.

'push'  y 'pop' se pueden implementar como métodos de la clase de pila como se muestra a continuación.

Adicionalmente, el método is_empty() que verifica si una pila está vacía también debe definirse como un método de clase de pila.

```python
class Stack:

  def __init__(self):
    self.stack=[]
  
  def is_empty(self):
    return True if len(self.stack)==0 else False

  def push(self,item):
    self.stack.append(item)

  def pop(self):
    return None if self.is_empty() else self.stack.pop()
```

**Líneas 1,3,6,7,9,10,12,13:**

- 1: Defina la estructura de datos de la pila como una clase de pila.
- 3: El constructor __init__(self) inicializa el campo miembro de la pila en una lista vacía.
- 6-7: El método is_empty() devuelve True si la longitud de self.stack es 0, y si no, devuelve False.
- 9: El método push() recibe el 'self' y el 'item' que se agregarán a la pila como parámetros.
- 10: El 'push' agrega el item a la parte superior de la lista self.stack, que es el campo miembro de la clase de pila.
- 12: El método pop() recibe 'self' como parámetro.
- 13: El 'pop' devuelve None si self.is_empty() es True, y si no, elimina y devuelve el elemento superior de self.stack.

El siguiente código prueba la estructura de datos de la pila que se definió como una clase anteriormente.

```python
stack=Stack()
stack.push('A')
stack.push('B')
print(stack.pop())
stack.push('C')
print(stack.pop())
print(stack.pop())
print(stack.pop())
```

**Líneas 1-8:**

La variable de pila guarda la instancia de objeto de una clase de pila que se creó llamando al constructor Stack().

¿Cómo implementaría la función **peek()**que le permite saber qué elemento está en la parte superior de la pila?

El método peek(), simplemente devuelva el elemento superior sin eliminar el elemento.

```python
def peek(stack):
  return None if is_empty(stack) else stack[-1]
```

También es posible definir el método peek() como un método miembro de la clase de pila que se definió previamente de la siguiente manera.

```python
def peek(self):
  return None if is_empty(self.stack) else self.stack[-1]
```

**TIP**

Al designar el valor del índice como -1 en la lista de Python, es posible acercarse al elemento superior de la lista.

stack[-1] == stack[len(stack) – 1]

```python
def check_parentheses(expr):
  opening= '[{('
  closing=']})'
  stack=Stack()
  out_result = True
  for char in expr:
    if char in opening:
      stack.push(char)
    elif char in closing:
      open_char = stack.pop()
      if open_char:
        if opening.index(stack.pop()) != closing.index(char):
          out_result = False
      else:
        out_result = False
  if not stack.is_empty():
    out_result = False
  return out_result
```

<html><body> <h1> hello world </h1>

```python
text= input('Ingrese la cadena del documento HTML: ')
```

```python
start=text.find('<')
while start != -1:
  end= text.find('>',start +1)
  tag=text[start:end+1]
  print(tag,end=' ')
  start=text.find('<',end+1)
```
