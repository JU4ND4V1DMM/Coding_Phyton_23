#### **2.1. Cómo funciona la solución**

El problema de la caída del huevo encuentra la altura donde se rompe el huevo al recibir la "altura" del edificio como entrada del usuario.

```python
def do_experiment(floor,breaking):
  return floor >= breaking

def find_highest_safe_floor(height,breaking):
  for n in range(1,height+1):
    if do_experiment(n,breaking):
      return n-1
  return height
```

```python
from random import randint

height=int(input('Ingrese el número de pisos: '))
breaking=randint(1,height)
floor=find_highest_safe_floor(height,breaking)
print(f'Su huevo estará a salvo hasta el piso #{floor}.')
```

#### **3.1. Dos tipos de búsqueda**

En programación, uno de los procesos más frecuentes es encontrar un elemento específico en unos datos dados.

Si los datos proporcionados no tienen ninguna regla, deben ser verificados todos. Por ejemplo, considere un problema para encontrar la ubicación de un valor aleatorio x de la lista S donde se enumeran aleatoriamente N enteros. ¿Qué pasa si x no existe en la lista S dada? No podrá encontrar la ubicación de x hasta que verifique todos los elementos en S.

Si los datos dados tienen reglas, el problema de búsqueda se puede resolver de manera más eficiente. Por ejemplo, si los datos dados son una secuencia ordenada, se puede usar la búsqueda binaria, que se discutirá más adelante.

##### **3.1.1. Búsqueda secuencial**

La búsqueda secuencial se refiere a un algoritmo que investiga comprobando todos los elementos uno por uno en orden.

La búsqueda secuencial se realiza para datos dados aleatoriamente. Entonces, debido a que la búsqueda se realiza sin reglas, también se conoce como "encontrar una aguja en un pajar".

#### **Un algoritmo para la búsqueda secuencial**

Considere un algoritmo de búsqueda secuencial que encuentra un número entero específico en una lista dada de números enteros.

A continuación se muestra el problema del algoritmo de búsqueda secuencial y su entrada y salida.

- Problema: ¿Existe un entero aleatorio x en un conjunto de enteros dados guardados en la lista S?
- Entrada: S es la lista de enteros, x como un entero aleatorio para encontrar.
- Salida: la ubicación del índice si x existe en la lista S, y -1 si no existe.

```python
def seq_search(nums,x):
  for i in range(len(nums)):
    if x==nums[i]:
      return i
  return -1
```

```python
S=[11,37,45,26,59,28,17,53]
x=53
pos=seq_search(S,x)
print(f'En la lista S, x={x} esta en la posición {pos}.')
```

El algoritmo de búsqueda secuencial compara el valor x para encontrar con cada uno de los elementos presentes la lista S dada.

Al iniciar la búsqueda secuencial, devuelve el índice 0 si el valor es igual al del primer elemento, y en caso contrario, realiza otra comparación con el siguiente elemento.

#### **Implementar la búsqueda secuencial**

```python
def seq_search(nums,x):
  for i in range(len(nums)):
    if x==nums[i]:
      return i
  return -1
```

#### **Un paso más**

¿Cuál es la complejidad temporal de la búsqueda secuencial? Analice con la notación Big-O.

Primero, en el mejor de los casos, la búsqueda finaliza después de una sola comparación, por lo que ciertamente es O(1). Además, se requiere la comparación de n veces en el peor de los casos, por lo que es claramente O (n).

¿Cómo se analiza la complejidad temporal promedio de la búsqueda secuencial? Suponga que S tiene N elementos. Al buscar cada elemento de S uno por uno, calcule cuántas operaciones de comparación (==) se deben realizar.

Asigne 1 al primer elemento y 2 al segundo, y así sucesivamente. Para el elemento N, se deben realizar operaciones de comparación N veces. Por lo tanto, el número de operaciones de comparación requeridas para buscar cada elemento una vez es N(N+1)/2.

\begin{align}
 1 + 2 + ...+ N=\frac{N(N+1)}{2}
 \end{align}


\


Entonces, el tiempo promedio de búsqueda T(n) para todos los elementos es O(n), que es un tiempo de búsqueda lineal.

 \begin{align}
T(N)=\frac{N(N+1)}{2}-N=\frac{(N+1)}{2} \epsilon  \  O(n)
 \end{align}

#### **4.1. Un algoritmo para encontrar la clave más grande**

Considere un algoritmo que encuentra la clave más grande en la lista de enteros dados.

La siguiente es la entrada y salida del problema del algoritmo para encontrar la clave más grande.

- Problema: ¿Dónde está la clave más grande en un conjunto de números enteros dados, S?
- Entrada: S como la lista de números enteros.
- Salida: Mayor índice entre los elementos de la lista S.

```python
def find_largest(nums):
  largest=0
  for i in range(1,len(nums)):
    if nums[largest]<nums[i]:
      largest=i
  return largest
```

```python
nums=[11,37,45,26,59,28,17,53]
pos=find_largest(nums)
print(f'El mayor elemento entre los elementos de la lista S es {nums[pos]} que esta en la posición {pos}')
```
