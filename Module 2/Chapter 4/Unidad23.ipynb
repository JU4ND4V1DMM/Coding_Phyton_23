#### **3.1. ¿Cómo resolver el problema de Josefo?**

Reciba N y K como entrada del usuario para imprimir la secuencia de Josefo.

```python
def josephus_sequence(n,k):
  sequence=[]
  queue=[]
  for i in range(1,n+1):
    queue.append(i)
  j=1
  while len(queue)>1:
    item=queue.pop(0)
    if j % k == 0:
      sequence.append(item)
    else:
      queue.append(item)
    j +=1
  sequence.append(queue.pop(0))
  return sequence
```

### <FONT COLOR="blue"> 4. ¿Qué es una cola (Queue)?</FONT>

#### **4.1. Definición de una Queue**

Una queue es un tipo de datos abstracto en el que solo es posible borrar en la parte delantera y solo agregar en la parte trasera.
Debido a que el primer elemento agregado también se elimina primero, una cola es una estructura de datos que funciona con el método FIFO (primero en entrar, primero en salir) (First-In-First-Out).


Una cola debe proporcionar dos interfaces importantes.

Las colas funcionan muy similar a como funcionan las listas, estas se componen de el ya clásico nodo y de dos métodos principales, los cuales son Enqueue que se encarga de encolar información, y el método Dequeue el cual se encarga de sacar la información de la cola y borrarla.


- **Enqueue** - poner en cola (datos): agrega un nuevo elemento a la fila de la cola. Si la cola está llena, entonces se dice que es una condición de desbordamiento - Complejidad de tiempo: O(n). El lado se llama trasero (rear).
- **Dequeue**: elimina un elemento del principio de la cola. Los elementos se abren en el mismo orden en que se empujan. Si la cola está vacía, entonces se dice que es una condición de subdesbordamiento - Complejidad de tiempo: O(1). El lado se llama delantera (front).

#### **5.1. Definir una queue como un tipo de clase**

En Python, el tipo de datos de lista se puede usar para definir la estructura de datos de la cola como una clase.

```python
class Queue:
  
  def __init__(self):
    self.queue =[]

  def is_empty(self):
    return True if len(self.queue)==0 else False
  
  def enqueue(self,item):
    self.queue.append(item)

  def dequeue(self):
    return None if self.is_empty() else self.queue.pop(0)
```

Similar a la clase de pila, la clase de cola define una cola como un tipo de datos de lista a través del método constructor.

#### **6.3. Implementación y codificación**

Implemente los pasos anteriores en los códigos de Python. Primero, la función josepush() recibe números enteros n y k como parámetros de entrada.

```python
class Queue():
  def __init__(self):
    self.queue =[]

  def is_empty(self):
    return True if len(self.queue)==0 else False
  
  def enqueue(self,item):
    self.queue.append(item)

  def dequeue(self):
    return None if self.is_empty() else self.queue.pop(0)
  
  def size(self):
    return len(self.queue)
```

```python
def josephus(n,k):
  sequence=[]
  queue=Queue()
  for i in range(1,n+1):
    queue.enqueue(i)
  j=1
  while queue.size() >1:
    item=queue.dequeue()
    if j % k == 0:
      sequence.append(item)
    else:
      queue.enqueue(item)
    j +=1
  sequence.append(queue.dequeue())
  return sequence
```

**Lineas 1-13:**

- 1-3: La función josephus() devuelve la permutación de Josephus en una lista a través de los parámetros de entrada n y k.
La sequence de retorno se inicializa como una lista vacía, mientras que la cola se inicializa como un objeto de la clase Queue que se definió previamente.
- 4-5: Poner en cola los enteros del 1 al N en la primera cola.
- 6-7,13: 'j' es un contador que saca un elemento de la cola. Por lo tanto, procesa la inicialización con 1.
Luego, aumente el valor de j en 1 para repetir poner y quitar de la cola hasta que solo quede un elemento en la cola.
Si el residuo de dividir j con k es 0, entonces este elemento es el siguiente k-ésimo que se eliminará de la cola.

- 8-12: Quitar un elemento de la cola.
Si j % k == 0, entonces es el k-ésimo elemento, así que agregue un elemento al final de la lista sequence.
Si j % k != 0, este elemento no se quitará de la cola, así que vuelva a agregarlo a la cola.
Repita el proceso hasta que solo quede un elemento en la cola y los elementos eliminados se guarden en la lista sequence.

- 14-15: Una vez que se completa el bucle while, solo queda un elemento en la cola, que se eliminará de la cola.
Por lo tanto, es posible imprimir la secuencia -sequence tomando el elemento de la cola y agregándolo al final de la secuencia.
