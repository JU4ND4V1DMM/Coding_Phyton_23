#### **2.1. Un algoritmo para la búsqueda binaria**

Considere el algoritmo de búsqueda binaria que encuentra un entero específico de la lista de enteros ordenados.

El siguiente es el problema del algoritmo de búsqueda binaria, y su entrada y salida.

- Problema: ¿Existe un entero aleatorio x en un conjunto de enteros ordenados, S?.
- Entrada: S como la lista de números enteros dispuestos en orden ascendente y x como el número entero aleatorio a encontrar.
- Salida: La ubicación del índice si x existe en la lista S, y -1 si no existe.

```python
def bin_search(nums,x):
  low,high=0, len(nums)-1
  while low<=high:
    mid=(low+high)//2
    if nums[mid]==x:
      return mid
    elif nums[mid]>x:
      high=mid-1
    else:
      low=mid+1
  return -1
```

```python
S=[11,17,26,28,37,45,53,59]
x=int(input('Ingrese el número a buscar: '))
pos=bin_search(S,x)
print(f'En la lista S, x={x} esta en la posición {pos}.')
```

El algoritmo de búsqueda binaria busca el valor x en la lista dada S.

El espacio de búsqueda inicial está entre el bajo y el alto (bajo como el índice del primer elemento y alto como el índice del último elemento) en S.

#### **2.2. Implementación y codificación**

El algoritmo de búsqueda binaria se puede implementar de la siguiente manera. 

Reciba 'nums' y x como entrada para configurar los nums completos como espacio de búsqueda.

```python
def bin_search(nums,x):
  low,high=0, len(nums)-1
  while low<=high:
    mid=(low+high)//2
    if nums[mid]==x:
      return mid
    elif nums[mid]>x:
      high=mid-1
    else:
      low=mid+1
  return -1
```

**Líneas 2-11:**

- 2:El 'low' y 'high' se inicializan como el primer y último elemento de los números, respectivamente.
En la parte inicial de la búsqueda binaria, los números enteros son el espacio de búsqueda. Por lo tanto, el espacio de búsqueda es [0, len(nums)-1].

La búsqueda binaria compara el valor medio que está entre low y high. Dado que los datos están ordenados, una sola comparación puede reducir a la mitad el espacio de búsqueda.

- 3-10: El valor medio se calcula como (low + high) // 2. Aquí, el operador // devuelve el cociente de la división.
Si el valor de nums[mid] es el mismo que x, la búsqueda es exitosa, así que devuelva el mid.
Si el valor de nums[mid] es mayor que x, cambie 'high' a mid-1, y si es menor que x, cambie 'low' a mid+1.
Si el valor low o high cambia a mid + 1 o mid – 1 , entonces el espacio de búsqueda se reduce a la mitad.

Devuelve -1 si x no se encuentra en el proceso de búsqueda binaria.

- 11: Si el valor low es mayor que el valor high, rompa la instrucción while.
Si es así, devuelva -1 porque el valor de x no existe en S.

#### **Un paso más**

¿Cuál es la complejidad temporal de la búsqueda binaria?

Primero, en el mejor de los casos puede terminar la búsqueda solo en la primera vez, por lo que es O (1). 

Entonces, ¿cómo analizaría la complejidad temporal de la búsqueda binaria en el peor de los casos?
Suponga que S tiene N elementos. El tamaño del primer espacio de búsqueda es N. El segundo espacio de búsqueda es la mitad del original, por lo que es N/2. El siguiente espacio de búsqueda se reduce a la mitad nuevamente, que es N/4. Esta explicación se puede generalizar de la siguiente manera.

\begin{align}
N,\frac{N}{2},\frac{N}{2^2},...,\frac{N}{2^k}
 \end{align}

La búsqueda binaria finaliza si solo hay un espacio de búsqueda, que es 

\begin{align}
\frac{N}{2^k}=1
 \end{align}

 

 \begin{align}
N=2^k
 \end{align}



\begin{align}
log_2N=k
 \end{align}





 El máximo número de operaciones de búsqueda binaria es log<sub>2</sub>(⁡𝑁), por lo que la complejidad temporal de la búsqueda binaria en el peor de los casos es Ο(log<sub>2</sub>(⁡𝑁)).
