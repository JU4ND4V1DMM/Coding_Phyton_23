- Ser capaz de procesar datos almacenados en formatos csv, excel y json.
- Pivotación y remodelación flexibles del conjunto de datos.
- Ventanas deslizantes con valores de índice, indexación y creación de subconjuntos de datos.
- Adición y eliminación de columnas.
- Funciones de división, aplicación y combinación para cambiar o combinar datos como una poderosa herramienta de agrupación de datos.
- Integración con combinación de un conjuntos de datos de alto rendimiento.

### <FONT COLOR="blue"> 6. Series </FONT>

Series es la estructura de datos básica de pandas.
Es similar a la matriz Numpy, pero es diferente porque la serie tiene índices. 

Como se muestra en la siguiente figura, el valor del índice y los datos tienen una relación de 1:1. 
serie.PNG

Por lo tanto, la estructura en sí es similar al diccionario que tiene una estructura {clave: valor}.
Cada valor de índice funciona como la dirección de cada dato.

```python
import pandas as pd
s=pd.Series([1,2,3,4,5])
print(s)
```

A partir del código de muestra proporcionado anteriormente, no se requiere la designación separada de un índice para crear series. Pandas genera automáticamente un índice de tipo entero que inicia desde 0 y aumenta en 1 para cada elemento de datos.

**TIP**

Los índices de las series se pueden designar de manera flexible. No necesariamente comienza desde 0. También se pueden crear índices que no sean de tipo entero.
Al crear series, se utiliza 
el parámetro **index** para designar los valores de índice deseados.

```python
import pandas as pd
s=pd.Series([1,2,3,4,5], index=['q',1,'cat','d','dog'])

print(s)
```

Visite el enlace https://pandas.pydata.org/docs/reference/series.html para verificar los parámetros disponibles para Series.

### <FONT COLOR="blue"> 9. Operaciones aritméticas básicas para objetos de serie (+, -, /, *, ...) </FONT>

Pandas realiza tres etapas integradas de procesos para operaciones aritméticas entre objetos de datos. Los datos se organizan en cada fila y columna, y los elementos en el mismo índice (ubicación) se emparejan. Por lo tanto, los elementos que no están emparejados se clasifican como NaN.

#### **9.1. Operación aritmética para objetos de una sola serie**

```python
import pandas as pd

s=[10,20,30,40,50]
sr=pd.Series(s)

print(sr)
print(sr**2)
```

Línea 7:

Todos los elementos de la serie sr se multiplican por 2.

#### **9.2. Operación con índice para objetos de dos series**

Si los elementos de cada índice no coinciden, el resultado de la operación devuelve NaN.

```python
import pandas as pd

s=[10,20,30,40,50]
t=[23,345,123]
sr1=pd.Series(s)
sr2=pd.Series(t)

sr3=sr1+sr2

print(sr3)
```

Línea 8:

Regresa la suma de las series sr1 y sr2.

#### **Dibujar un gráfico de barras**

Un gráfico de barras representa la altura proporcional al tamaño del valor de los datos como una barra rectangular. Es posible comparar el tamaño de los datos a través de la diferencia relativa entre la altura de la barra, y hay dos tipos de barra diferentes, incluida la barra vertical y la barra horizontal.

- **Barra vertical plt.bar()**

  - Adecuado para mostrar diferencias entre valores de datos de dos puntos que tienen diferencia horaria.
  - En otras palabras, es adecuado para la expresión de datos de series de tiempo.

\

- **Barra horizontal plt.barh()**

  - Adecuado para mostrar diferencias entre tamaños de valores variables.

```python
from matplotlib import pyplot as plt
days_in_year=[88,225,365,687,4333,10756,30687,60190,90553]
plt.bar(range(len(days_in_year)),days_in_year)
plt.show()
```

### <FONT COLOR="blue">Práctica de la programación en parejas </FONT>


**Directriz, mecanismos y plan de contingencia**

 Preparar la programación por parejas implica establecer pautas y mecanismos para ayudar a los alumnos a emparejarse correctamente y mantenerlos emparejados.  Por ejemplo, los estudiantes deben turnarse para "conducir el mouse".  La preparación efectiva requiere planes de contingencia en caso de que uno de los socios se ausente o decida no participar por una u otra razón.  En estos casos, es importante dejar claro que el alumno activo no será sancionado porque el emparejamiento no funcionó bien.
 
**Emparejar habilidades similares, no necesariamente iguales, como socios**

 La programación en parejas puede ser efectiva cuando los estudiantes con habilidades similares, aunque no necesariamente iguales, se emparejan como compañeros.  El emparejamiento de estudiantes que no coinciden a menudo puede conducir a una participación desequilibrada.  Los maestros deben enfatizar que la programación en parejas no es una estrategia de "divide y vencerás", sino más bien un verdadero esfuerzo de colaboración en cada esfuerzo por todo el proyecto.  Los maestros deben evitar emparejar a estudiantes muy débiles con estudiantes muy fuertes.

**Motivar a los estudiantes ofreciendo incentivos adicionales**

 Ofrecer incentivos adicionales puede ayudar a motivar a los estudiantes a emparejarse, especialmente con estudiantes avanzados.  A algunos maestros les ha resultado útil pedirles a los estudiantes que se emparejen solo para una o dos tareas.

**Evitar trampas en la colaboración**

 El desafío para el maestro es encontrar formas de evaluar los resultados individuales, mientras aprovecha los beneficios de la colaboración.  ¿Cómo saber si un estudiante aprendió o hizo trampa?  Los expertos recomiendan revisar el diseño y la evaluación del curso, así como discutir explícita y concretamente con los estudiantes los comportamientos que se interpretarán como trampa.  Los expertos alientan a los maestros a hacer que las tareas sean significativas para los estudiantes y a explicar el valor de lo que los estudiantes aprenderán al completarlas.

**Entorno de aprendizaje colaborativo**

 Un entorno de aprendizaje colaborativo ocurre cada vez que un instructor requiere que los estudiantes trabajen juntos en actividades de aprendizaje.  Los entornos de aprendizaje colaborativo pueden implicar actividades tanto formales como informales y pueden incluir o no una evaluación directa.  Por ejemplo, parejas de estudiantes trabajan en asignaciones de programación; pequeños grupos de estudiantes discuten posibles respuestas a la pregunta de un profesor durante la lección; y los estudiantes trabajan juntos fuera de clase para aprender nuevos conceptos.  El aprendizaje colaborativo es distinto de los proyectos en los que los estudiantes "dividen y vencerán".  Cuando los estudiantes dividen el trabajo, cada uno es responsable de solo una parte de la resolución del problema y hay oportunidades muy limitadas para resolver los problemas con los demás.  En entornos colaborativos, los estudiantes participan en conversaciones intelectuales entre ellos.
