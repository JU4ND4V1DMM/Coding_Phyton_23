DataFrame es una estructura de datos de etiqueta bidimensional compuesta de filas y columnas. (Las series se diferencian de un DataFrame debido a que éstas son matriz unidimensional).

### <FONT COLOR="blue"> 2. Creando un Dataframe </FONT>

Se combinan varias matrices unidimensionales de la misma longitud para formar un DataFrame. La misma longitud significa que el número de elementos de datos es el mismo. En otras palabras, se pueden combinar varias Series para crear un DataFrame.
Además, dado que un objeto que combina varias Series es un diccionario, también puede entenderse como "convertir un diccionario en un DataFrame".

#### **2.1. Conceptos básicos para la creación de un Dataframe**

**pandas.DataFrame**[El objeto de diccionario que se desea convertir].

```python
import pandas as pd
data={'col_1':[3,2,1,0],'col_2':['a','b','c','d']}
pd.DataFrame(data)
```

#### **2.2. Establecer el nombre de la fila (index) / columnas (column) del DataFrame**

Se pueden establecer los nombres de la fila y columna por separado al crear un DataFrame o después de crearlo.

Primero, se debe comprender la estructura y las reglas de nomenclatura de DataFrames a través de las imágenes presentadas a continuación.


dataframe3.PNG

Como se describe en la imagen, cada columna de un DataFrame es un objeto de serie, y estos objetos de serie tienen una estructura de matriz donde se combinan según el índice de la misma fila.

A continuación se presenta cómo se pueden cambiar los nombres de los índices de las filas y los nombres de las columnas:

- Cambiar todos los índices de fila: **DataFrame object name.index** = Matriz de índices de fila para cambiar.

- Cambiar todos los nombres de columna (nombre de columna): 
  
  **DataFrame object name.columns** = Matriz de nuevos nombres de columna.

```python
import pandas as pd
d={'col':[1,2],'col2':[3,4]}
df=pd.DataFrame(data=d)
print(df)

df.index=['new index0','new index1']
print(df)

df.columns=['new_col1','new_col2']

print(df)
```

**Líneas 2,3,4,6,7,9,11:**

- 2: Es un tipo de datos de diccionario.
- 3: Crea un objeto DataFrame con el nombre df asignando un diccionario.
- 4: Imprime el DataFrame creado antes del cambio de nombre.
- 6: Cambiar a nuevo nombre de índice.
- 7: Comprobar el resultado del cambio de índice.
- 9: Cambiar al nuevo nombre de columna.
- 11: comprobar el resultado del cambio de nombre de columna.


Como puede ver en el resultado de la ejecución del código anterior, cuando cambia el índice o el nombre de la columna, puede ver que el DataFrame anterior al cambio no se mantiene, pero el DataFrame se cambia con el índice y el nombre de la columna modificados.

Adicionalmente, se puede seleccionar una parte de una fila y cambiarle el nombre: **DataFrame object name.rename**(index={índice existente: nuevo nombre de índice, ...})

Igualmente al seleccionar parte de una columna se le puede cambiar el nombre: **DataFrame object name.rename**(columns={Nombre de columna existente: nuevo nombre de columna, ...})

```python
import pandas as pd
d={'col':[1,2],'col2':[3,4]}
df=pd.DataFrame(data=d)
print(df)

df.rename(index={0:'new index0'})
```

**Líneas 4,6:**

- 4: Verifica el DataFrame creado antes del cambio de nombre.
- 6: Cambia el índice 0 a un nuevo nombre que es 'new index0'.

```python
df.rename(columns={'col2':'new col2'})
```

**Línea 1:**

Cambia el nombre de la columna col2 por el nuevo nombre que es new col2.


Se puede observar que el resultado de ejecutar el código anterior, es diferente de cuando se cambió todo el índice y el nombre de las columnas. Es decir, que está devolviendo un nuevo objeto DataFrame en lugar de cambiar el original en sí. Se debe usar **inplace=True** para cambiar el objeto original.

```python
import pandas as pd
d={'col':[1,2],'col2':[3,4]}
df=pd.DataFrame(data=d)
df.rename(index={0:'new index0'},inplace=True)
df.rename(columns={'col2':'new col2'})
```

**Líneas 4,5:**

- 4: Realiza cambios en el objeto DataFrame original a través de la opción inplace = True.
- 5: Aunque solo se ha cambiado el nombre de la columna, se puede visualizar que el objeto original en sí se ha cambiado a través del código de la línea anterior.

#### **3.2. Acerca de las rutas de archivo**

Uno de los muchos errores que se cometen como principiantes mientras se aprende Python para el procesamiento de datos en primer lugar es ingresar la ruta incorrecta a los archivos para cargar datos.

Puede dejarlo como lo básico de la computación, pero es útil asegurarse de limpiar las rutas de los archivos una vez. Las rutas de los archivos en la computadora local se pueden expresar de dos maneras:

- **Ruta absoluta:** este método utiliza todas las rutas desde el primer punto de inicio (inicio del sistema operativo) hasta el archivo.

  Ejemplo: El sistema operativo utiliza el sistema Windows. Si encuentra "sample.txt" en el escritorio, es C:\Users\UserID\Desktop\sample.txt. No importa qué sistema operativo (SO) se use además de Windows, esto es para encontrar el archivo con una ruta absoluta que contenga todas las rutas pasadas desde la raíz de nivel superior.

- **Ruta relativa:** Es la "ubicación del archivo que desea encontrar" en función de "la ubicación actual".

  - / : Mover al directorio de nivel superior (raíz).
  - ./ : Mismo directorio actual, se puede eliminar.
  - ../ : Directorio principal desde mi ubicación actual (archivo en creación).
  - ../../ : Directorio dos niveles más arriba.


path.PNG


**¿Por qué necesitamos rutas relativas?**

Una ruta absoluta es una cadena estática que le dice exactamente dónde se encuentra un archivo en una computadora específica. Sin embargo, cuando se trata de rutas, esta característica estática puede ser una desventaja.

Ejemplo: ¿Qué sucede si la ruta de test.txt si se esta cambiando frecuentemente o si el directorio raíz cubre diferentes sistemas operativos? Debido a las características estáticas, el primero necesita reescribir todos los documentos escritos con rutas absolutas y el segundo debe crear y administrar rutas absolutas para cada sistema operativo.

##### **4.1.2. Opción Skiprows**

Especifica el número de filas iniciales que se van a omitir antes de cargar los datos. También se puede configurar como una lista que contiene el número de líneas a omitir.

El valor predeterminado es 0 (ninguno). Después de que el sistema tenga un registro binario candidato de una fila de entrada, determina si se inserta registro en la tabla de destino:
Si no ha especificado esta opción, el sistema inserta todos los registros.
Si ha especificado esta opción y el contador de filas de entrada es menor o igual que el recuento de skipRows, el sistema descarta el registro binario candidato (omitido). De lo contrario, el sistema inserta el registro.

Ejemplo [1,4,7]

```python
titanic=pd.read_csv("sample_data/test.csv")

titanic.head(5)
```

```python
titanic=pd.read_csv("sample_data/test.csv",skiprows=2)

titanic.head(5)
```

#### **6.1. JSON**


JSON (Notación de objetos de JavaScript - JavaScript Object Notation) es un formato creado para compartir datos. Consiste en pares atributo-valor o pares clave-valor. Un objeto JSON es un formato de datos de clave-valor que normalmente se representa entre llaves. 

Los pares clave-valor tienen dos puntos entre ellos, como en "key": "value". Cada par clave-valor está separado por una coma, por lo que el centro de un JSON suele ser así: "key": "value", "key": "value", "key": "value". 



**¿Cómo usarse?: "Nombre de datos- Key": valor**

Las claves JSON se encuentran en el lado izquierdo de los dos puntos. Deben estar entre comillas dobles, como en "key", y pueden ser cualquier cadena válida. Dentro de cada objeto, las claves deben ser únicas. Estas cadenas de teclas pueden incluir espacios en blanco, como en "first name", pero esto puede dificultar el acceso al programar, por lo que es mejor usar guiones bajos, como en "first_name". Los valores JSON se encuentran a la derecha de los dos puntos.

Contiene cadenas, números, matrices, valores booleanos y otros objetos, que son los tipos de datos básicos de JavaScript. De esta manera,los datos se pueden configurar como se muestra en la imagen del siguiente ejemplo.

- Si se crea como un archivo, la extensión del archivo es ".json"
- El tipo de contenido de la solicitud HTTP es "aplicación/json"

Esta es una función que calcula la correlación por pares de columnas excluyendo los valores NA/Nulo. Los tres métodos siguientes se pueden utilizar para esta función. Entre los datos de la mayoría de las columnas utilizadas por Pearson, se encuentran y comparan los datos numéricos, y los datos de tipo string  se restan naturalmente.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html

- pearson : coeficiente de correlación estándar.
- kendall : Coeficiente de correlación Tau de Kendall.
- Spearman: correlación de rango de Spearman.

El resultado se devuelve como un DataFrame y la interpretación de los valores es la siguiente.

- Cercano a 1, ambos aumentan por igual.
- Cercnano a -1, aumentar en uno / disminuir en uno.
- Cercano a 0, no hay relación entre los dos.

**seaborn.heatmap()**

Un mapa de calor es una representación gráfica de los valores contenidos en una matriz mediante el uso de colores. En estos, los índices de la matriz representan las variables de dos características que se desean comparar. Mientras que el color utilizado en cada elemento de la matriz representa la magnitud de la relación existente. Siendo una herramienta excelente para para mostrar las relaciones existentes entre las variables de diferentes características, ya que al mostrar la relación mediante un color se obtiene una interpretación fácil e intuitiva de esta. No solo esto, sino que al mismo tiempo se puede ver comparar las relaciones de unas características con otras observando otros puntos del mapa de calor.

Parametrós más utilizados en esta función:

- center: El valor en el cual centrar el mapa de color al representar los datos.
- cmap: Indica el mapa que se utilizará para la representación de los valores.
- annot: Indica si se representa o no la magnitud de cada celda en el mapa además del color, por defecto no se representará.
- fmt: Es el formato con el que se representará la magnitud.

### <FONT COLOR="blue">Práctica de la programación en parejas </FONT>


**Directriz, mecanismos y plan de contingencia**

 Preparar la programación por parejas implica establecer pautas y mecanismos para ayudar a los alumnos a emparejarse correctamente y mantenerlos emparejados.  Por ejemplo, los estudiantes deben turnarse para "conducir el mouse".  La preparación efectiva requiere planes de contingencia en caso de que uno de los socios se ausente o decida no participar por una u otra razón.  En estos casos, es importante dejar claro que el alumno activo no será sancionado porque el emparejamiento no funcionó bien.
 
**Emparejar habilidades similares, no necesariamente iguales, como socios**

 La programación en parejas puede ser efectiva cuando los estudiantes con habilidades similares, aunque no necesariamente iguales, se emparejan como compañeros.  El emparejamiento de estudiantes que no coinciden a menudo puede conducir a una participación desequilibrada.  Los maestros deben enfatizar que la programación en parejas no es una estrategia de "divide y vencerás", sino más bien un verdadero esfuerzo de colaboración en cada esfuerzo por todo el proyecto.  Los maestros deben evitar emparejar a estudiantes muy débiles con estudiantes muy fuertes.

**Motivar a los estudiantes ofreciendo incentivos adicionales**

 Ofrecer incentivos adicionales puede ayudar a motivar a los estudiantes a emparejarse, especialmente con estudiantes avanzados.  A algunos maestros les ha resultado útil pedirles a los estudiantes que se emparejen solo para una o dos tareas.

**Evitar trampas en la colaboración**

 El desafío para el maestro es encontrar formas de evaluar los resultados individuales, mientras aprovecha los beneficios de la colaboración.  ¿Cómo saber si un estudiante aprendió o hizo trampa?  Los expertos recomiendan revisar el diseño y la evaluación del curso, así como discutir explícita y concretamente con los estudiantes los comportamientos que se interpretarán como trampa.  Los expertos alientan a los maestros a hacer que las tareas sean significativas para los estudiantes y a explicar el valor de lo que los estudiantes aprenderán al completarlas.

**Entorno de aprendizaje colaborativo**

 Un entorno de aprendizaje colaborativo ocurre cada vez que un instructor requiere que los estudiantes trabajen juntos en actividades de aprendizaje.  Los entornos de aprendizaje colaborativo pueden implicar actividades tanto formales como informales y pueden incluir o no una evaluación directa.  Por ejemplo, parejas de estudiantes trabajan en asignaciones de programación; pequeños grupos de estudiantes discuten posibles respuestas a la pregunta de un profesor durante la lección; y los estudiantes trabajan juntos fuera de clase para aprender nuevos conceptos.  El aprendizaje colaborativo es distinto de los proyectos en los que los estudiantes "dividen y vencerán".  Cuando los estudiantes dividen el trabajo, cada uno es responsable de solo una parte de la resolución del problema y hay oportunidades muy limitadas para resolver los problemas con los demás.  En entornos colaborativos, los estudiantes participan en conversaciones intelectuales entre ellos.
