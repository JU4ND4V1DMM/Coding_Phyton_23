### <FONT COLOR="blue"> 1. ¿Por qué es necesario el ordenamiento de datos? </FONT>

Hasta ahora, hemos aprendido a crear DataFrames y series, o traer otros archivos y convertirlos en un DataFrame. De hecho, la situación más frecuente cuando se realizan tareas similares es cuando la forma de datos del objetivo del análisis es anormal.
Los datos anormales se refieren a casos en los que faltan datos, las unidades no coinciden o se superponen sin reglas. El proceso de organización de los datos anómalos se denomina ordenamiento de datos.

Este término fue acuñado por Hadley Wickham (http://hadley.nz/) en su artículo "Tidy Data" (Link del paper: https://vita.had.co.nz/papers/tidy-data.html) y el documento está disponible para descargarse.

tidy data.PNG

Hay algo que los científicos de datos en el campo siempre dicen.
**"El 80% del análisis de datos se dedica a ordenar y preparar datos".**
Además, la razón por la que este trabajo es tan tedioso y repetitivo es porque la preparación y ordenamiento de datos no termina en un solo proceso, sino que requiere limpiar repetidamente nuevos problemas que se encuentran constantemente en el proceso.
Por lo tanto, Tidy Data proporciona un método estándar para configurar valores de datos dentro de un conjunto de datos estandarizados. El estándar facilita la limpieza de los datos iniciales porque no tiene que empezar desde cero y realizar nuevas tareas cada vez.

tidy data2.PNG

https://www.openscapes.org/blog/2020/10/12/tidy-data/

### <FONT COLOR="blue">2. Definición de datos ordenados </FONT>

“Los datos ordenados son una forma estándar de asignar el significado de un conjunto de datos a su estructura” (Hadley Wickham).

**Las 3 reglas de Tidy Data**

- Cada variable debe tener su propia columna.
- Cada observación debe tener su propia fila.
- Cada valor debe tener su propia celda.

tidy data3.PNG

### <FONT COLOR="blue"> 3. Datos desordenados </FONT>

El siguiente es un resumen de situaciones reales en las que es necesario ordenar los datos.

**Datos desordenados:**

- Si la columna no es el nombre de la variable, sino el valor en sí.
- Si no hay una sola variable en la columna, sino multiples variables.
- Si las variables se almacenan tanto en columnas como en filas (deben almacenarse en columnas).
- Si existen datos faltantes.
- Si no es una muestra para el período de tiempo deseado.
- Si se necesitan datos cuantitativos pero la variable es cualitativa.
- Si los tipos de datos de los valores son incorrectos.
- Si hay duplicidad de datos.

### <FONT COLOR="blue"> 8. Reemplazar datos faltantes con otros datos </FONT>

**DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)**

Eliminar todas las columnas o filas es desfavorable solo porque faltan datos en la etapa de preprocesamiento del conjunto de datos,  porque reduce la cantidad de conjuntos de datos que se analizarán.
Sin embargo, reemplazar los datos de NaN con valores como 0 o 1 afectará el análisis de datos.
Por lo tanto, generalmente es un valor que reemplaza los datos faltantes, y se obtiene el valor medio y el valor de la moda, que representan la distribución y las características del conjunto de datos.

#### **8.4. Reemplazar con valores vecinos más cercanos**

Esto no es cierto para todos los datos, pero por lo general, debido a su naturaleza, los datos vecinos a menudo tienen similitudes. Después de verificar las características del conjunto de datos, los datos faltantes se reemplazan con datos anteriores o inmediatamente posteriores.

- Cambiar el índice de fila completo: **dataframeobjectname.index = matriz de índices de fila que se van a cambiar**.
- Cambiar el nombre completo de la columna: **dataframeobject.columns = matriz de nombres de columna que se cambiarán**.

**Línea 1:**

**Varianza:**

Es posible verificar si los datos están dispersos o concentrados alrededor de la media calculando la varianza. Después de designar la variable de referencia, se utiliza el método **var()**.

La Varianza es una medida de dispersión que se utiliza para representar la variabilidad de un conjunto de datos respecto de la media aritmética de los mismo. Así, se calcula como la suma de los residuos elevados al cuadrado y divididos entre el total de observaciones

**Desviación estándar:**

Cuando hay una gran cantidad de datos, el promedio se usa a menudo como el valor que representa los datos. 

La desviación estándar, en uno de los diagramas de dispersión, es una figura representativa que indica qué tan dispersos están los datos alrededor del promedio. La unidad de la desviación estándar es idéntica a la unidad de datos, a diferencia de la varianza que usa la raíz cuadrada. Si la desviación estándar está cerca del centro, significa que los valores de los datos están concentrados cerca del promedio. Cuanto mayor sea la desviación estándar, más generalizados estarán los valores de los datos.

### **Paso 4:**

**Coeficiente de variación (CV):**

Los datos sobre variables con diferentes unidades de medida no se pueden comparar simplemente. Esto se debe a que si el tamaño de los datos es diferente, la desviación tiende a aumentar cuando la unidad de medida es grande.

Ejemplo: 

La desviación estándar entre los precios de las acciones y los precios del gas no se puede comparar simplemente.

Lo que podemos usar en estas situaciones es el coeficiente de la variación.
Es el valor de la desviación estándar dividido por la media. Esto se puede utilizar para comparar datos de diferentes especificaciones independientemente del tamaño.

**Coefficient of Variation (CV) = Standard Deviation/Mean**

```python
study_time_cv=student_data['studytime'].std()/student_data['studytime'].mean()
print(study_time_cv)
```

```python
absences_cv=student_data['absences'].std()/student_data['absences'].mean()
print(absences_cv)
```

### <FONT COLOR="blue">Práctica de la programación en parejas </FONT>


**Directriz, mecanismos y plan de contingencia**

 Preparar la programación por parejas implica establecer pautas y mecanismos para ayudar a los alumnos a emparejarse correctamente y mantenerlos emparejados.  Por ejemplo, los estudiantes deben turnarse para "conducir el mouse".  La preparación efectiva requiere planes de contingencia en caso de que uno de los socios se ausente o decida no participar por una u otra razón.  En estos casos, es importante dejar claro que el alumno activo no será sancionado porque el emparejamiento no funcionó bien.
 
**Emparejar habilidades similares, no necesariamente iguales, como socios**

 La programación en parejas puede ser efectiva cuando los estudiantes con habilidades similares, aunque no necesariamente iguales, se emparejan como compañeros.  El emparejamiento de estudiantes que no coinciden a menudo puede conducir a una participación desequilibrada.  Los maestros deben enfatizar que la programación en parejas no es una estrategia de "divide y vencerás", sino más bien un verdadero esfuerzo de colaboración en cada esfuerzo por todo el proyecto.  Los maestros deben evitar emparejar a estudiantes muy débiles con estudiantes muy fuertes.

**Motivar a los estudiantes ofreciendo incentivos adicionales**

 Ofrecer incentivos adicionales puede ayudar a motivar a los estudiantes a emparejarse, especialmente con estudiantes avanzados.  A algunos maestros les ha resultado útil pedirles a los estudiantes que se emparejen solo para una o dos tareas.

**Evitar trampas en la colaboración**

 El desafío para el maestro es encontrar formas de evaluar los resultados individuales, mientras aprovecha los beneficios de la colaboración.  ¿Cómo saber si un estudiante aprendió o hizo trampa?  Los expertos recomiendan revisar el diseño y la evaluación del curso, así como discutir explícita y concretamente con los estudiantes los comportamientos que se interpretarán como trampa.  Los expertos alientan a los maestros a hacer que las tareas sean significativas para los estudiantes y a explicar el valor de lo que los estudiantes aprenderán al completarlas.

**Entorno de aprendizaje colaborativo**

 Un entorno de aprendizaje colaborativo ocurre cada vez que un instructor requiere que los estudiantes trabajen juntos en actividades de aprendizaje.  Los entornos de aprendizaje colaborativo pueden implicar actividades tanto formales como informales y pueden incluir o no una evaluación directa.  Por ejemplo, parejas de estudiantes trabajan en asignaciones de programación; pequeños grupos de estudiantes discuten posibles respuestas a la pregunta de un profesor durante la lección; y los estudiantes trabajan juntos fuera de clase para aprender nuevos conceptos.  El aprendizaje colaborativo es distinto de los proyectos en los que los estudiantes "dividen y vencerán".  Cuando los estudiantes dividen el trabajo, cada uno es responsable de solo una parte de la resolución del problema y hay oportunidades muy limitadas para resolver los problemas con los demás.  En entornos colaborativos, los estudiantes participan en conversaciones intelectuales entre ellos.
