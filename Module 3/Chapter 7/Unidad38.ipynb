#### **5.3. Crea intervalos de tiempo con Pandas Timedelta**

- Hemos aprendido a crear un momento de fecha y hora hasta ahora. Ahora, aprenderemos a expresar intervalos de tiempo. timedelta es una subclase de datetime.timedelta de Python y también es compatible con Pandas.
- Puede usar la clase Timedelta de Pandas para intervalos de tiempo. El intervalo de tiempo es importante para el análisis de datos de series temporales porque es necesario para determinar el número de días o analizar por un intervalo de tiempo específico.

``` pandas.Timedelta(value, unit=None, **kwargs) ```

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Timedelta.html

````python
import pandas as pd
import datetime
from datetime import datetime 
# Define tu fecha de nacimiento
my_birth = datetime(1973,7,7)
# Calcula la fecha de hoy
today = datetime.today()
# Calcula la fecha de mañana
tomorrow = today + pd.Timedelta(days = 1)
# Calcula la diferencía entre mañana y tu fecha de nacimiento
print(tomorrow-my_birth)
````

````python
# ¿Que objeto de python se retorna cuando se restan las fechas?
tomorrow-my_birth
````

````python
today = datetime.today()
example_date = today + pd.Timedelta(hours=10)
print(f"la datetime actual en el servidor de colab es {today}")
print(f"la datetime dentro de 10 hroas en el servidor de colab es {example_date}")
````

#### **5.4. Representar un período con Period**

**pandas.Period()**

https://pandas.pydata.org/docs/reference/api/pandas.Period.html?highlight=pd%20period

La mayor parte del análisis de datos de series temporales es un análisis de eventos para un intervalo de tiempo específico. Un ejemplo es un análisis de las ventas de una empresa durante un período de tiempo específico. Sin embargo, cuando se analizan eventos agrupando varios períodos, es difícil usar solo marcas de tiempo. Pandas proporciona un intervalo de tiempo estandarizado a través de una clase llamada **Period** para facilitar este tipo de organización y cálculo de datos.
Period crea un período basado en una frecuencia específica, como diaria, semanal, mensual, anual, trimestral, etc. y proporciona una marca de tiempo que indica la hora de inicio y la hora de finalización.
Se puede crear un período utilizando una marca de tiempo correspondiente a un punto de referencia y una frecuencia que indica el período.
Si crea un período correspondiente a un mes basado en julio de 1973, puede hacerlo de la siguiente manera.

````python
import pandas as podemos

special_day=pd.Period('1973-7',freq='M')
special_day
````

pd.Period tiene atributos que muestran la hora de inicio (start_time) y la hora de finalización (end_time).

https://pandas.pydata.org/docs/reference/api/pandas.Period.start_time.html#pandas.Period.start_time

````python
special_day.start_time
````

**Línea 1:**

Devuelve la hora de inicio de ese punto en el tiempo.

````python
special_day.end_time
````

**Línea 1:**

Devuelve la hora de fin de ese punto en el tiempo.

El período se puede cambiar a través de una operación aritmética simple. Puede crear un nuevo objeto de período cambiando la frecuencia. El siguiente ejemplo es un ejemplo de +2 (desplazamiento de dos meses) porque la frecuencia de special_day es de un mes.

````python
n=special_day + 2

print(special_day)
print(n)
````

#### **7.1. Crear series de tiempo de frecuencia específica a partir de la matriz de marca de tiempo**

Los datos de series de tiempo se pueden crear no solo en unidades de una unidad, sino también en un intervalo de tiempo específico, es decir, en una forma con una frecuencia específica.
Puede usar el parámetro **freq** para** pd.date_range()** para crear una serie temporal con la frecuencia que desee. El valor predeterminado es una unidad.
El principio es similar a crear una matriz de números con Python **range()** .


**`pandas.date_range(start=None, end=None, periods=None, freq=None, tz=None, normalize=False, name=None, closed=None, **kwargs)`**


- **start:** el inicio del rango de fechas, 
- **end:** el final del rango de fechas, 
- **periods:** el número de marcas de tiempo que se crearán.

Para obtener más detalles sobre la frecuencia, consultar:

 https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases


Ejemplos representativos de frecuencia:

- D: frecuencia de días calendario.
- B: día hábil (Frecuencia de días hábiles).
- W: frecuencia semanal.
- M: frecuencia de fin de mes.
- Ms: frecuencia de inicio del mes.
- Q: frecuencia final de cuarto.

### <FONT COLOR="blue"> 8. Función de estadísticas móviles (cálculos de ventana móvil) </FONT>

Pandas proporciona una función que puede calcular fácilmente las estadísticas móviles (ventana móvil) para DataFrame y Series.

La expresión “ventana” se refiere a un período o tramo en el que se expresan datos específicos.

Es un formato en el que la ventana se mueve automáticamente según el intervalo especificado, las estadísticas se calculan en consecuencia y se aplican todos los datos de la serie temporal.

Para explicar haciendo referencia a la imagen a continuación, el promedio móvil es el método más utilizado para el análisis de datos financieros de series de tiempo para suavizar la volatilidad a corto plazo de los datos objetivo y analizar tendencias a largo plazo.

rolling window.PNG
https://docs.wavefront.com/query_language_windows_trends.html


**TIP**

Suavizado significa un procesamiento uniforme mediante la eliminación de pequeñas fluctuaciones o discontinuidades que no son buenas en los datos debido al ruido cuando se muestrean datos de muestras grandes.


- Función móvil en DataFrame: 

**`pandas.DataFrame.rolling()`**

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html


- Función móvil en Serie: 

**`pandas.Series.rolling()`**

https://pandas.pydata.org/docs/reference/api/pandas.Series.rolling.html?highlight=series%20rolling#pandas.Series.rolling

Practiquemos los datos del precio de las acciones usando el concepto de promedio móvil como ejemplo

**Método representativo**

rolling window2.PNG

````python
import pandas as pd
import datetime
from datetime import date, datetime, time 
import matplotlib.pyplot as plt

import pandas_datareader.data as web
start=datetime(2020,1,1)
end=datetime(2020,12,31)

sec = web.DataReader('005930.KS',
                   data_source='yahoo',
                   start=start, 
                   end=end)
sec.head()
````

**Líneas 6,7,10,11,12,13:**
- 6: Es un conjunto de datos de acciones de Corea proporcionado por Yahoo Finance en el lector de datos.
- 7: Si desea practicar con una ventana amplia, es mejor especificar un período más largo.

- 13: busque la fecha de finalización, utilizando la marca de tiempo almacenada en la variable de finalización

````python
sec.info()
````

**Línea 1:**

Puede ver que este conjunto de datos ya está en DatetimeIndex.

````python
ma5=sec['Adj Close'].rolling(window=5).mean()
ma5.head(10)
````

**Línea 1:**

Es fácil calcular los datos del promedio móvil durante 5 días.

Visualicemos cada dato de promedio móvil

````python
close=sec['Adj Close']

ma5=sec['Adj Close'].rolling(window=5).mean()
ma10=sec['Adj Close'].rolling(window=10).mean()
ma60=sec['Adj Close'].rolling(window=60).mean()
ma120=sec['Adj Close'].rolling(window=120).mean()

plt.figure(figsize=(15,10))

plt.plot(close, label='close', linewidth=4)
plt.plot(ma5, label='5 window')
plt.plot(ma10, label='10 window')
plt.plot(ma60, label='60 window')
plt.plot(ma120, label='120 window')

plt.legend()
plt.title('Rolling window calculations')
plt.xlabel('Date')
plt.ylabel('close price')
````

La anterior figura es el gráfico de resultados.

### <FONT COLOR="blue">Práctica de la programación en parejas </FONT>


**Directriz, mecanismos y plan de contingencia**

 Preparar la programación por parejas implica establecer pautas y mecanismos para ayudar a los alumnos a emparejarse correctamente y mantenerlos emparejados.  Por ejemplo, los estudiantes deben turnarse para "conducir el mouse".  La preparación efectiva requiere planes de contingencia en caso de que uno de los socios se ausente o decida no participar por una u otra razón.  En estos casos, es importante dejar claro que el alumno activo no será sancionado porque el emparejamiento no funcionó bien.
 
**Emparejar habilidades similares, no necesariamente iguales, como socios**

 La programación en parejas puede ser efectiva cuando los estudiantes con habilidades similares, aunque no necesariamente iguales, se emparejan como compañeros.  El emparejamiento de estudiantes que no coinciden a menudo puede conducir a una participación desequilibrada.  Los maestros deben enfatizar que la programación en parejas no es una estrategia de "divide y vencerás", sino más bien un verdadero esfuerzo de colaboración en cada esfuerzo por todo el proyecto.  Los maestros deben evitar emparejar a estudiantes muy débiles con estudiantes muy fuertes.

**Motivar a los estudiantes ofreciendo incentivos adicionales**

 Ofrecer incentivos adicionales puede ayudar a motivar a los estudiantes a emparejarse, especialmente con estudiantes avanzados.  A algunos maestros les ha resultado útil pedirles a los estudiantes que se emparejen solo para una o dos tareas.

**Evitar trampas en la colaboración**

 El desafío para el maestro es encontrar formas de evaluar los resultados individuales, mientras aprovecha los beneficios de la colaboración.  ¿Cómo saber si un estudiante aprendió o hizo trampa?  Los expertos recomiendan revisar el diseño y la evaluación del curso, así como discutir explícita y concretamente con los estudiantes los comportamientos que se interpretarán como trampa.  Los expertos alientan a los maestros a hacer que las tareas sean significativas para los estudiantes y a explicar el valor de lo que los estudiantes aprenderán al completarlas.

**Entorno de aprendizaje colaborativo**

 Un entorno de aprendizaje colaborativo ocurre cada vez que un instructor requiere que los estudiantes trabajen juntos en actividades de aprendizaje.  Los entornos de aprendizaje colaborativo pueden implicar actividades tanto formales como informales y pueden incluir o no una evaluación directa.  Por ejemplo, parejas de estudiantes trabajan en asignaciones de programación; pequeños grupos de estudiantes discuten posibles respuestas a la pregunta de un profesor durante la lección; y los estudiantes trabajan juntos fuera de clase para aprender nuevos conceptos.  El aprendizaje colaborativo es distinto de los proyectos en los que los estudiantes "dividen y vencerán".  Cuando los estudiantes dividen el trabajo, cada uno es responsable de solo una parte de la resolución del problema y hay oportunidades muy limitadas para resolver los problemas con los demás.  En entornos colaborativos, los estudiantes participan en conversaciones intelectuales entre ellos.
